```python
import matplotlib.pyplot as plt


col = [
    'Transaction ID', 'Item', 'Quantity', 'Price Per Unit',
    'Total Spent', 'Payment Method', 'Location', 'Transaction Date'
]


total_cells = df[col].size


nan_count = df[col].isna().sum().sum()

error_count = (
    df[col]
    .astype(str)
    .apply(lambda c: c.str.lower().eq("error"))
    .sum()
    .sum()
)

unknown_count = (
    df[col]
    .astype(str)
    .apply(lambda c: c.str.lower().eq("unknown"))
    .sum()
    .sum()
)

# Results dictionary
results = {
    "Valid": total_cells - nan_count - error_count - unknown_count,
    "Missing": nan_count + error_count + unknown_count
}

# Print stats
print(f"Total cells: {total_cells}")
print(f"Nulls: {nan_count} ({nan_count / total_cells * 100:.2f}%)")
print(f"ERRORs: {error_count} ({error_count / total_cells * 100:.2f}%)")
print(f"UNKNOWNs: {unknown_count} ({unknown_count / total_cells * 100:.2f}%)")

# Pie chart
plt.figure(figsize=(5, 5))
plt.pie(
    results.values(),
    labels=results.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#ff9999']
)
plt.title('Data Quality Distribution Across Entire Dataset', fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Here i cant drop all nulls due to its high percentage 

```python
df=df.dropna(subset='Item')
df = df[~df['Item'].astype(str).str.lower().isin(['error','unknown'])]
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce") 
df["Quantity"] = df["Quantity"].fillna(df.groupby("Item")["Quantity"].transform("median"))
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
```

```python
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce") 
df["Price Per Unit"] = df["Price Per Unit"].fillna(df.groupby("Item")["Price Per Unit"].transform("median"))
df["Price Per Unit"] = df["Price Per Unit"].fillna(df["Price Per Unit"].median())
```

```python
import numpy as np
df['Total Spent']=df['Total Spent'].replace(['UNKNOWN','ERROR'],np.nan)
df['Total Spent']=df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

```python
df.head(50)
```

```python
df['Payment Method']=df['Payment Method'].replace(['UNKNOWN','ERROR'],np.nan)
```

```python
df['Location']=df['Location'].replace(['UNKNOWN','ERROR'],np.nan)
```

```python
df['Transaction Date']=df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors="coerce")
```

```python
import matplotlib.pyplot as plt


col = [
    'Transaction ID', 'Item', 'Quantity', 'Price Per Unit',
    'Total Spent', 'Payment Method', 'Location', 'Transaction Date'
]


total_cells = df[col].size


nan_count = df[col].isna().sum().sum()

error_count = (
    df[col]
    .astype(str)
    .apply(lambda c: c.str.lower().eq("error"))
    .sum()
    .sum()
)

unknown_count = (
    df[col]
    .astype(str)
    .apply(lambda c: c.str.lower().eq("unknown"))
    .sum()
    .sum()
)

# Results dictionary
results = {
    "Valid": total_cells - nan_count - error_count - unknown_count,
    "Missing": nan_count + error_count + unknown_count
}

# Print stats
print(f"Total cells: {total_cells}")
print(f"Nulls: {nan_count} ({nan_count / total_cells * 100:.2f}%)")
print(f"ERRORs: {error_count} ({error_count / total_cells * 100:.2f}%)")
print(f"UNKNOWNs: {unknown_count} ({unknown_count / total_cells * 100:.2f}%)")

# Pie chart
plt.figure(figsize=(5, 5))
plt.pie(
    results.values(),
    labels=results.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#ff9999']
)
plt.title('Data Quality Distribution Across Entire Dataset', fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
